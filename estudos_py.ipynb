```python
# Validação da Nota
while True:
    try:
        nota = float(input("Digite sua nota (0 a 10): "))
        if 0 <= nota <= 10:
            break
        else:
            print("❌ Erro: A nota deve estar entre 0 e 10.")
    except ValueError:
        print("⚠️ Erro: Digite apenas números para a nota!")

# Validação da Frequência
while True:
    try:
        frequencia = float(input("Digite sua frequência (0 a 100): "))
        if 0 <= frequencia <= 100:
            break
        else:
            print("❌ Erro: A frequência deve estar entre 0 e 100.")
    except ValueError:
        print("⚠️ Erro: Digite apenas números para a frequência!")

# Lógica de aprovação
if nota >= 5 and frequencia >= 70:
    print("Aprovado")
elif nota >= 5 and frequencia < 70:
    print("Reprovado por frequencia sem direito a P3")
elif 4 <= nota < 5 and frequencia >= 70:
    print("Recuperação")
elif nota < 4 and frequencia >= 70:
    print("Reprovado por nota")
else:
    print("Reprovado")
```
